```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(start=0, stop=1, num=100)  #para revisar en la siguiente line pon solo la "x" y run it
```

```python
def my_sin(x):
    return np.sin(x)
def my_cos(x):
    return np.cos(x)
    
```

```python
fig, axes = plt.subplots(1,2)
axes[0].plot(x, my_sin(x))
axes[1].plot(x, my_cos(x))
axes[0].set_xlabel("sin curve")
axes[1].set_xlabel("cos curve")
fig.suptitle("Prompt 8")
plt.savefig("Prompt8.pdf", dpi=300)
plt.show()
```
